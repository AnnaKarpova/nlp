```python
ngram_sorted = data.ngram.str.lower().str.translate(None, ''.join(['%'])).str.lstrip().str.rstrip()
```

```python
data_sorted=pd.concat([data.drop('ngram',1), ngram_sorted], axis=1)
```

```python
data_sorted=data_sorted[data_sorted.ngram!='']
len(data_sorted)
```

```python
ngram_list=[t[0] for t in Counter(data_sorted.ngram).most_common()]
ngram_list[0]
```

```python
indexes=[]
corpus_id=[]
ngram_id=[]
for index, row in data_sorted.iterrows():
    indexes.append(index)
    corpus_id.append(corpus_list.index(row['corpus']))
    ngram_id.append(ngram_list.index(row['ngram']))
    
```

```python
ngram_id_df=pd.Series(data=ngram_id, index=indexes,name='ngram_id')
corpus_id_df=pd.Series(data=corpus_id, index=indexes,name='corpus_id')
```

```python
data_ids=pd.concat([corpus_id_df, data_sorted['docid'], ngram_id_df, data_sorted['count']], axis=1)
```

```python
data_ids.head(5)
```

```python
doc_length=defaultdict(int)
for index,row in data_ids.iterrows():
    if row['docid'] in doc_length.keys():
        doc_length[row['docid']]+=row['count']
    else:
        doc_length[row['docid']]=row['count']
```

```python
doc_length
```

```python
indexes=[]
doc_length_list=[]
for index,row in tqdm(data_ids.iterrows()):
    indexes.append(index)
    doc_length_list.append(doc_length[row['docid']])
    
```

```python
doc_length_df=pd.Series(data=doc_length_list, index=indexes,name='doc_length')
```

```python
data_fin=pd.concat([data_ids, doc_length_df], axis=1)
```

```python
data_fin.head(5)
```

```python
len(data_fin['docid'].unique())
```

```python
data_fin_sorted=data_fin.sort_values(by=['doc_length','docid'], ascending=[False, True])
```

```python
data_fin_sorted=data_fin_sorted.reset_index(drop=True)
```

```python
data_fin_sorted.head(5)
```

```python
docid_unique=data_fin_sorted.docid.drop_duplicates().tolist()
docid_test=[]
docid_train=[]
for i in xrange(len(docid_unique)):
    if i%10!=9:
        docid_train.append(docid_unique[i])
    else:
        docid_test.append(docid_unique[i])
        
```

```python
data_test=data_fin_sorted[data_fin_sorted['docid'].isin(docid_test)]
data_train=data_fin_sorted[data_fin_sorted['docid'].isin(docid_train)] 
```
